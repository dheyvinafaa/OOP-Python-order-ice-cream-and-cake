```python
import matplotlib.pyplot as plt 
import pandas as pd
df = pd.read_csv("/content/Apartemen_numerik.csv")
print(df)
```

```python
fig = plt.figure(figsize=(5,10))
ax = plt.axes()
ax.pie(df.KodeApt, labels = df.KodeApt)
plt.title("KodeApt"
plt.show()
```
